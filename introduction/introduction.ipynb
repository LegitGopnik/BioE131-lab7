# FASTA format
FASTA is the most common format for nucleotide or aminoacid sequences.
More information [here](https://en.wikipedia.org/wiki/FASTA_format).

```python
# Let's look at an example file
# cat command prints the contents of the file
!cat example.fa
```

```python
# Reading a fasta file in Python

from Bio import SeqIO

for seq_rec in SeqIO.parse("example.fa", "fasta"):
    print(seq_rec.name)
    print(seq_rec.seq)
```

# Simple plots with pandas and seaborn

```python
# When importing matplotlib, convention is to rename it “plt” as in:
import matplotlib.pyplot as plt

# To get iPython to plot stuff in-line, this magic command needs to go somewhere:
%matplotlib inline
```

```python
import pandas as pd
import seaborn as sns

# loading an example dataset
penguins = sns.load_dataset("penguins")
penguins
```

```python
# scatter plot
sns.scatterplot(data=penguins, x="bill_length_mm", y="flipper_length_mm", hue="species");
```

```python
# box plot
sns.boxplot(data=penguins, x="species", y="body_mass_g");
```

```python
# histogram
sns.histplot(data=penguins, x="flipper_length_mm", hue="species");
```

```python
# stacked bar plot
penguins.groupby("island").sex.value_counts(normalize=True).unstack().plot.bar(stacked=True);
```

```python
# another stacked bar plot
penguins.groupby("species").island.value_counts(normalize=True).unstack().plot.bar(stacked=True);
```

# Plotting phylogenetic trees with BioPython

We will work with phylogenetic trees in [Newick format](https://evolution.genetics.washington.edu/phylip/newicktree.html).

```python
# Let's inspect the first tree. cat command prints the contents of the file
!cat tree1.nkw
```

```python
from Bio import Phylo
```

```python
# Let's read the tree from the file and plot it
tree1 = Phylo.read("tree1.nkw", "newick")
Phylo.draw(tree1)
```

```python
# Same topology but specifying branch lengths
!cat tree2.nkw
```

```python
tree2 = Phylo.read("tree2.nkw", "newick")
Phylo.draw(tree2)
```

```python
# Changing figure size
fig = plt.figure(figsize=(8, 4))
Phylo.draw(tree2, axes=fig.gca())
```
